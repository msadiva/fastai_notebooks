```python
import fastai 
from fastai.vision.all import *
import numpy as np
import os
```

```python
help(ImageDataLoaders.from_name_func)
```

```python
def is_cat(x):
  return x[0] == "c"
```

```python
path = os.getcwd() + "/train/"
```

```python
path
```

```python
dls = ImageDataLoaders.from_name_func(path, get_image_files(path), valid_pct= 0.25, seed = 42, label_func = is_cat, item_tfms = Resize(224))
```

```python
help(cnn_learner)
```

```python
learn = cnn_learner(dls,resnet18, metrics = error_rate )
```

```python
learn.fine_tune(3)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(1, nrows = 1, figsize = (5, 5))
```

```python
dls.valid.show_batch(max_n = 8, nrows = 2)
```
